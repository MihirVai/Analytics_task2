```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import clear_output
```

```python
df=pd.read_csv("data.csv", )
```

```python
satis=df.loc[:, "Engagement Score (% Satisfaction)"]
satis=np.array(satis)
salary=df.loc[:, "Monthly Income"]
salary=np.array(salary)
age=df.loc[:, "Age"]
age=np.array(age)
Tenure=df.loc[:, "Tenure"]
Tenure=np.array(Tenure)
Gender=df.loc[:, "Gender"]
designation=df.loc[:, "Designation"]
for i in range(np.size(satis)):
  satis[i]=float(satis[i].strip('%'))
for i in range(np.size(Tenure)):
  Tenure[i]=float(Tenure[i].replace(" ", ""))
    
```

```python
plt.figure(figsize=(20, 5))
plt.scatter(satis, salary)
plt.show()
```

**This graph shows that satisfaction rating dosent affect the rate at which people are leaving as the scatter is all over the x axis, however employees with higher income are less likely to leave than those with lower income**

```python
fig=plt.figure(figsize=(25,4))
plt.scatter(Tenure, Gender )
```

**This graph shows that men are in much larger numbers than women, and that 6 units of tenure is a threshhold for resigning as most people have left before that threshhold**

```python

plt.scatter(age, designation)
```

**45 years of age is a threshhold for leaving the company as majority of the employees have left by that age. Further Sr Sales Executive, Sales Executive are the two main designations from where resignaions occur**

```python
plt.figure()
df['Marital Status'].value_counts().plot()
plt.show()

plt.figure()
df['Remarks'].value_counts().plot(figsize=(20,4))
plt.show()
```

**These graphs show that divorced people leave more frequently than the single and the married as well as that issues with the manager is the most common reason for leaving followed by more challenging job roles**
